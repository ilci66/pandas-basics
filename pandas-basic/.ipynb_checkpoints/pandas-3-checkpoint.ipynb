```python
# figsize(float, float), default: rcParams["figure.figsize"] (default: [6.4, 4.8])
axs = air.plot.area(figsize=(12, 4), subplots=True)
```

```python
axs = air.plot.area(figsize=(12, 4), subplots=False)
```

**Separate subplots for each of the data columns are supported by the subplots argument of the plot functions.**

```python
# For futher customization, extending or saving the result
fig, axs = plt.subplots(figsize=(12, 4))

air.plot.area(ax=axs)
```

```python
axs.set_ylabel("NO$_2$ concentration")
```

```python
# saving the plot, in the folder I'm currently working 
fig.savefig("no2_concentrations.png")
```



    The .plot.* methods are applicable on both Series and DataFrames

    By default, each of the columns is plotted as a different element (line, boxplot,)

    Any plot created by pandas is a Matplotlib object.
